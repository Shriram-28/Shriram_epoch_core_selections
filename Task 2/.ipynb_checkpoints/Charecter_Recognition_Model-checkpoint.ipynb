```python
def reverse_dict(original_dict):
  """Reverses the keys and values in a dictionary and returns a new dictionary.

  Args:
      original_dict (dict): The original dictionary to be reversed.

  Returns:
      dict: A new dictionary with keys and values reversed.
  """
  reversed_dict = {}
  for key, value in original_dict.items():
    # Check if the value already exists as a key (handling duplicates)
    if value in reversed_dict:
      reversed_dict[value].append(key)  # Add key to a list if value already exists
    else:
      reversed_dict[value] = key  # Set the value as the key and the key as the value (single occurrence)
  return reversed_dict
```

```python
from tensorflow.keras import layers
from tensorflow.keras import regularizers
model = Sequential()

# Convolutional layers with ReLU activation and padding for consistency
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1), kernel_regularizer=regularizers.l2(0.001)))
model.add(MaxPool2D(pool_size=(2, 2), strides=2))

model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu', padding='same', kernel_regularizer=regularizers.l2(0.001)))
model.add(MaxPool2D(pool_size=(2, 2), strides=2))

model.add(Conv2D(filters=128, kernel_size=(3, 3), activation='relu', padding='valid', kernel_regularizer=regularizers.l2(0.001)))
model.add(MaxPool2D(pool_size=(2, 2), strides=2))

# Dropout for regularization
model.add(Dropout(0.2))  # Experiment with dropout rate (e.g., 0.2, 0.3, 0.5)

# Flatten layer
model.add(Flatten())

# Dense layers with ReLU activation and L2 regularization
model.add(Dense(64, activation="relu", kernel_regularizer=regularizers.l2(0.001)))
model.add(Dropout(0.2))  # Experiment with dropout rate

model.add(Dense(128, activation="relu", kernel_regularizer=regularizers.l2(0.001)))
model.add(Dropout(0.2))  # Experiment with dropout rate

# Output layer with softmax activation
model.add(Dense(26, activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer = Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(train_X_n, train_yOHE, epochs=20,  validation_data = (test_X_n,test_yOHE))
```

## Model-2 with slightly changed Architecture without Regularization 

```python
#downsampling the values
X_train = train_X/255.
X_test = test_X/255.
```

```python
from keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPooling2D
from keras.callbacks import EarlyStopping, ModelCheckpoint

model_2 = Sequential()

#First Conv1D layer
model_2.add(Conv2D(32,kernel_size = (3,3),activation = 'relu',input_shape = (28,28,1)))
model_2.add(MaxPooling2D(pool_size = (2,2),strides = 2))

#Second Conv1D layer
model_2.add(Conv2D(filters = 64, kernel_size = (3,3),activation = 'relu', padding = 'same'))
model_2.add(MaxPooling2D(pool_size = (2,2), strides = 2))

#Third Conv1D layer
model_2.add(Conv2D(filters = 128, kernel_size = (3,3),activation = 'relu', padding = 'valid'))
model_2.add(MaxPooling2D(pool_size = (2,2), strides = 2))

#Flatten layer
model_2.add(Flatten())

#Dense layer 1
model_2.add(Dense(128, activation = 'relu'))
model_2.add(Dropout(0.2))

#Dense layer 2
model_2.add(Dense(64,activation = 'relu'))

#Final layer of 26 nodes
model_2.add(Dense(26,activation = 'softmax'))

#Define the loss function to be categorical cross-entropy since it is a multi-classification problem:
model_2.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

### Early stopping and model checkpoints are the callbacks to stop training the neural network at the right time and to save the best model after every epoch:

```python
#Callbacks
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=10, min_delta=0.001) 
mc = ModelCheckpoint('Models/4_7_24/Callback models/Dropout Model/best_model.keras', monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
callbacks_list = [es,mc]
```

```python
history_2 = model_2.fit(X_train, train_yOHE, epochs=200, callbacks = callbacks_list , validation_data = (X_test,test_yOHE))
```

```python
#evaluating model on test dataset
model.evaluate(X_test,test_yOHE)
```

```python
import cv2
import os

def load_images_from_folder(folder_path):
  """
  This function loads all images from a folder and saves them in a list.

  Args:
      folder_path (str): The path to the folder containing the images.

  Returns:
      list: A list containing the loaded images as OpenCV image arrays.
  """

  images = []
  for filename in os.listdir(folder_path):
    # Check if the file is an image
    if filename.lower().endswith((".jpg", ".jpeg", ".png")):
      # Get the full path of the image
      image_path = os.path.join(folder_path, filename)
      # Load the image using cv2.imread
      image = cv2.imread(image_path)
      
      # Check if image loading is successful
      if image is not None:
        images.append(image)
      else:
        print(f"Error loading image: {filename}")

  return images
```
