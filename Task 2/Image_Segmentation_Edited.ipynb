```python
import cv2
import numpy as np

def split_image_grid(image, rows, cols):
  """Splits an image into a grid of separate images.

  Args:
      image (np.ndarray): Input image as a NumPy array.
      rows (int): Number of rows in the grid.
      cols (int): Number of columns in the grid.

  Returns:
      list: List of NumPy arrays representing the individual grid images.
  """
  # Get image height, width, and calculate cell dimensions
  height, width = image.shape[:2]
  cell_height, cell_width = height // rows, width // cols

  # Split the image into sub-images based on grid cells
  grid_images = []
  for i in range(rows):
    for j in range(cols):
      y_start = i * cell_height
      y_end = y_start + cell_height
      x_start = j * cell_width
      x_end = x_start + cell_width
      grid_image = image[y_start:y_end, x_start:x_end]
      grid_images.append(grid_image)

  return grid_images
```

```python
def display_grid_images(grid_images, title="Grid Images"):
  """Displays a grid of images in separate windows with titles.

  Args:
      grid_images (list): List of NumPy arrays representing the grid images.
      title (str, optional): Title for the display window. Defaults to "Grid Images".
  """
  # Create windows for each grid image
  for i, grid_image in enumerate(grid_images):
    cv2.imshow(f"{title} - Cell {i+1}", grid_image)

  # Wait for key press to close windows
  cv2.waitKey(0)
  cv2.destroyAllWindows()
```

```python
import cv2

def is_blank_image(image):
  """Checks if an image is mostly blank (black).

  Args:
      image (np.ndarray): Input image as a NumPy array.

  Returns:
      bool: True if the image is mostly blank, False otherwise.
  """
  # Convert to grayscale if needed
  if len(image.shape) == 3:
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  # Calculate the sum of all pixel values
  pixel_sum = image.sum()

  # Define a threshold for blank image (adjust based on image size)
  threshold = image.size * 0.01  # Consider 1% non-zero pixels as non-blank

  return pixel_sum < threshold
```

```python
import cv2
import os

def save_word_m(output_folder, word, word_count):
  """Saves a list of character images as a single word image."""
  for i, char_image in enumerate(word):
    filename = f"word_{word_count}"
    if i > 0:
      filename += f"_letter_{i+1}"  # Add suffix for subsequent characters
    cv2.imwrite(os.path.join(output_folder, filename + ".jpg"), char_image.astype(np.uint8))

def separate_letters_m(image_list, output_folder):
  """Separates letters from blank images and groups them word-wise.

  Args:
      image_list (list): List of NumPy arrays representing the input images (shape: (28, 28, 3) for color).
      output_folder (str): Path to the output folder for separated images.
  """

  # Create the output folder if it doesn't exist
  os.makedirs(output_folder, exist_ok=True)

  # Initialize variables
  blank_indexes = []  # List to store indexes of blank images
  current_word = []  # List to store images in the current word
  word_count = 0

  # Identify blank image indexes and group letter images
  for i, image in enumerate(image_list):
    # Convert image to grayscale (assuming color images)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to separate foreground (letter) from background (black)
    thresh, binary_image = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)  # Updated for OpenCV 4 compatibility

    # Check if the image is mostly empty (blank)
    if is_blank_image(grid_images[i]) == True:
      blank_indexes.append(i)
      if len(current_word) > 0:  # Save the previous word if any
        save_word_m(output_folder, current_word, word_count)
        current_word = []
      word_count += 1  # Increment word count for blank image (separator)
    else:
      current_word.append(image)  # Add letter image to current word

  # Save the last word if any characters were found
  if len(current_word) > 0:
    save_word_m(output_folder, current_word, word_count)
```

```python
import os
import cv2

def hconcat_and_save_images(image_groups, output_folder):
  """
  This function takes a dictionary of image groups (keys are numbers, values are image paths)
  and concatenates all images in each group horizontally (hconcat), saving them in a new folder.

  Args:
      image_groups (dict): A dictionary where keys are numbers and values are lists of image paths.
      output_folder (str): The path to the new folder for saving concatenated images.
  """

  os.makedirs(output_folder, exist_ok=True)  # Create output folder if it doesn't exist

  for number, image_paths in image_groups.items():
    # Validate if there are at least 2 images for concatenation
    if len(image_paths) < 2:
      print(f"Skipping group {number} - Less than 2 images for hconcat.")
      continue
    if len(image_paths) > 9:
      first_items = image_paths[:len(image_paths) - 9]  # Get first (length-9) items
      remaining_items = image_paths[len(image_paths) - 9:]  # Get remaining items
      image_paths = remaining_items + first_items # Combine with first items at the end
    # Load images as NumPy arrays (assuming compatible formats)
    images = []
    for image_path in image_paths:
      image = cv2.imread(image_path)
      images.append(image)

    # Horizontally concatenate the images
    combined_image = cv2.hconcat(images)

    # Create output filename with the number
    output_filename = f"{output_folder}/group_{number}.jpg"

    # Save the concatenated image
    cv2.imwrite(output_filename, combined_image)

# Example usage
folder_path = "groups_from_line_2"  # Replace with your actual folder path
 # Replace with your desired output folder name


hconcat_and_save_images(lis, folder_path)
```
